```python
#W10A3
def square_sum_digits(n):
    s=0
    while(n>0):
        s+=(n%10)**2
        n//=10;
    return s;
n=0
try:
    n=int(input())
    track=set()
    while True:
        if n==1:
            print("YES")
            break
        elif n in track:
            print("NO")
            break;
        else:
            track.add(n)
            n=square_sum_digits(n)
except ValueError:
    print("n is not a positive integer")


```
