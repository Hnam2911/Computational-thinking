```python
#W5A1
def Max_Of_Two(a,b):
    return a if a>b else b
a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
#W5A2
def swap(a,b):
    a,b=b,a
    return a,b
a,b=map(int,input().split())
print(swap(a,b))
```

```python
#W5A3
def prime(n):
    if n<2: return False
    if n==2 or n==3: return True
    for i in range(2,int(n**0.5+1)):
        if n%i==0:
            return False
    return True
n=int(input())
print(prime(n))
```

```python
#W5A4
def perfect_num(n):
    s=1
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            s+=i
            if n//i!=i:
                s+=(n//i)
    return s==n
n=int(input())
print(perfect_num(n))
```

```python
#W5A5
def find(arr,k):
    for i in range(1,len(arr)-1):
        if arr[i]==k:
            return i+1
    return -1
arr=list(map(int,input().strip().split()))
k=int(input())
print(find(arr,k))
```

```python
#W5A6
def factorial(n):
    s=1
    for i in range(2,n+1):
        s*=i
    return s
n=int(input())
print(factorial(n))
```

```python
#W5A7
def calculate(a,b,x):
    a,b=float(a),float(b)
    match x:
        case '+':
            return a+b
        case '-':
            return a-b
        case '*':
            return a*b
        case '/':
            return a/b
        
a,x,b=input().strip().split()
print(calculate(a,b,x))
```

```python
#W5A8
def binary(n):
    res=''
    while n>0:
        res+=str(n%2)
        n//=2
    return res[::-1]

def hamming_distance(a,b):
    a=binary(a)[::-1]
    b=binary(b)[::-1]
    while len(a)<len(b): a+='0'
    while len(b)<len(a): b+='0'
    c=0
    for i in range(len(a)):
        if a[i]!=b[i]:
            c+=1
    return c
a,b=map(int,input().split())
print(hamming_distance(a,b))
```

```python
#W5A9
def sum_of_digits(n):
    s=0
    while n>0:
        s+=(n%10)
        n//=10
    return s

n=int(input())
print(sum_of_digits(n))
```

```python
#W5A10
def check(a,b):
    if len(a)!=len(b): return False
    dp=[-1]*26
    for i in range(len(a)):
        id1,id2=ord(a[i])-96,ord(b[i])-96
        if id1==id2 or (dp[id1]==id2 and dp[id2]==id1):
            continue
        else:
            if dp[id1]==dp[id2]==-1:
                dp[id1],dp[id2]=id2,id1
            else:
                return False
    return True
            
a,b=input().split()
print(check(a,b))
                
                
                
            
    
```
