```python
#W6A3
arr=list(map(int,input().strip().split()))
k=int(input())
n=len(arr)
k%=n
print(k)
res=[-1]*n
for i in range(n):
    res[i]=arr[(i+k)%n]
res=tuple(res)
print(res)
```

```python
#W6A5
arr=list(map(int,input().split()))
dic={}
dic['positives']=0
dic['negatives']=0
dic['zeros']=0
for i in arr:
    if i>0:
        dic['positives']+=1
    elif i<0:
        dic['negatives']+=1
    else:
        dic['zeros']+=1
print(dic)
```

```python
#W6A9
s1=list(input().rstrip('}').lstrip("{").split(', '))
s2=list(input().rstrip('}').lstrip("{").split(', '))
dic={}
for i in s1:
    k,v=i.split(':')
    dic[k]=int(v)
for i in s2:
    k,v=i.split(':')
    dic[k]=dic.get(k,0)+int(v)
dic=dict(sorted(dic.items()))
print(dic)
```

```python
#W6A10
arr=list(map(int,input().split()))
k=int(input())
res=[]
for i in arr:
    if k-i in arr:
        tmp=(i,k-i) if i<k-i else (k-i,i)
        if tmp not in res:
            res.append(tmp)
print(res)
```

```python
#W6A12
arr=list(map(int,input().split()))
dic={}
for i in arr:
    dic[i]=dic.get(i,0)+1
res=0
max_c=0
for i in dic.keys():
    if dic[i]>max_c:
        res=i
        max_c=dic[i]
    elif dic[i]==max_c:
        if res>i:
            res=i
print(res)
```
