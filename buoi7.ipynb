```python
#W7A1
def binary_search(arr,k):
    l,r=0,len(arr)-1
    while l<=r:
        mid=(l+r)//2
        if arr[mid]==k:
            return mid
        elif arr[mid]>k:
            r=mid-1
        else:
            l=mid+1
    return -1
arr=list(map(int,input().split()))
k=int(input())
print(binary_search(arr,k))
```

```python
#W7A3
def selection_sort(arr):
    n=len(arr)
    for i in range(n-1):
        min_idx=i
        for j in range(i+1,n):
            if arr[min_idx]>arr[j]:
                min_idx=j
        arr[i],arr[min_idx]=arr[min_idx],arr[i]
def bubble_sort(arr):
    n=len(arr)
    for i in range(n-1):
        for j in range(i+1,n):
            if arr[j]<arr[i]:
                arr[j],arr[i]=arr[i],arr[j]
arr=list(map(int,input().split()))
selection_sort(arr)
print(arr)
```

```python
#W7A4
def max_fre(arr):
    dic={}
    for i in arr:
        dic[i]=dic.get(i,0)+1
    res=0
    max_fre=0
    for k in dic.keys():
        if max_fre<dic[k]:
            max_fre=dic[k]
            res=k
    return res,max_fre
arr=list(map(int,input().split()))
print(max_fre(arr))
```

```python
#W7A5
def pairs(arr,x):
    n=len(arr)
    dic={}
    c=0
    for i in arr:
        a=x-i
        if a in dic:
            c+=dic[a]
        dic[i]=dic.get(i,0)+1
    return c
arr=list(map(int,input().split()))
x=int(input())
print(pairs(arr,x))
```

```python
#W7A7
import bisect
def LIS(arr):
    tail=[]
    pos=[]
    pre=[-1]*len(arr)
    for i in range(len(arr)):
        idx=bisect.bisect_right(tail,arr[i])
        if idx==len(tail):
            tail.append(arr[i])
            pos.append(i)
        else:
            tail[idx]=arr[i]
            pos[idx]=i
        if idx>0: pre[i]=pos[idx-1]
    res=[]
    cur=pos[-1]
    while cur!=-1:
        res.append(arr[cur])
        cur=pre[cur]                
    return len(tail)
arr=list(map(int,input().split()))
print(LIS(arr))
```

```python
#W7A9
import bisect
def count(arr):
    n=len(arr)
    tmp=[]
    res=[0]*n
    for i in range(n-1,-1,-1):
        idx=bisect.bisect_left(tmp,arr[i])
        if idx==len(res):
            tmp.append(arr[i])
        else:
            tmp.insert(idx,arr[i])
        res[i]+=idx
    return res
arr=list(map(int,input().split()))
print(count(arr))

        
```

```python
#W7A10
def max_sum_modulo(arr,k):
    arr=[i%k for i in arr]
    n=len(arr)
    res=0
    for i in range(n):
        s=0
        for j in range(i,n):
            s+=arr[i]
            s%=k
            if s==k-1:
                return k-1
            if s>res:
                res=s
    return res

arr=list(map(int,input().split()))
k=int(input())
print(max_sum_modulo(arr,k))
        
```

```python
#W7A11
def count_word(s,w):
    res=0
    j=0
    point=[]
    n=len(s)
    i=0
    while i <n:
        if s[i]==w[0] and (not(s[i-1].isalpha()) or i==0):
            c=0
            j=0
            while i<n and j<len(w) and s[i]==w[j]:
                i+=1
                j+=1
                c+=1
            if c==len(w) and (i==n or not(s[i].isalpha())):
                res+=1
        else:
            i+=1
    return res
            
s=input()
w=input()
print(count_word(s,w))
```
