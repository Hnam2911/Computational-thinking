```python
#W4A6
def is_prime(b):
    is_prime=[1]*(b+1)
    is_prime[0]=is_prime[1]=0
    i=2
    while i*i<=b:
        if is_prime[i]:
            for j in range(i*i,b+1,i):
                is_prime[j]=0
        i+=1
    return is_prime
a,b=map(int,input().split())
s=0
check=is_prime(b)
for i in range(a,b+1):
    if check[i]:
        s+=i
print(s)
```

```python
#W4A8
def reverse(n):
    res=0
    while n>0:
        res=res*10+n%10
        n//=10
    return res
def palindrome(n):
    c=0
    while n!=reverse(n):
        n+=reverse(n)
        c+=1
    return f'{c} {n}'

n=int(input())
print(palindrome(n))
```

```python
#W4A9
def check(n):
    i=int(n**0.5)
    if i*i!=n: return False
    digit=[]
    for i in str(n):
        if i not in digit:
            digit.append(i)
        else:
            return False
    return True
        
n=int(input())
for i in range(1,n+1):
    if check(i):
        print(i,end=' ')
```

```python
#W4A32
def strong_pass(s):
    if len(s)<6: return False
    lst=[0]*4
    for i in s:
        if i.isdigit(): lst[2]=1
        elif i.isupper(): lst[1]=1
        elif i.islower(): lst[3]=1
        elif i!=' ': lst[0]=1
    return not 0 in lst

s=input()
print(strong_pass(s))
```
