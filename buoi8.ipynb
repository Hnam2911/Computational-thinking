```python
#W8A1
import math
pi=3.14
class Cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def total_area(self):
        return 2*pi*((self.r)**2)+2*(self.r)*pi*self.h
    def volume(self):
        return (self.r**2)*pi*self.h
r,h=map(int,input().split())
cy1=Cylinder(r,h)
print(cy1.volume())
print(cy1.total_area())
```

```python
#W8A3
class Point:
    def __init__(self,x,y) -> None:
        self.x=x
        self.y=y
    def check_0(self):
        return self.x==self.y==0
    def check_x_axis(self):
        return self.x==0
    def check_y_axis(self):
        return self.y==0
    def dis(self,a):
        res=((self.x-a.x)**2+(self.y-a.y)**2)**0.5
        return f'{res:.2f}'
x,y=map(int,input().split())
p1=Point(x,y)
print(p1.check_0())
print(p1.check_x_axis())
print(p1.check_y_axis())
print(p1.dis(Point(0,0)))
```

```python
#W8A4
class Calculator:
    def __init__(self,a,b) -> None:
        self.a=a
        self.b=b
    def add(self):
        return self.a+self.b
    def subtract(self):
        return self.a-self.b
    def divide(self):
        if self.b!=0:
            return self.a/self.b
        else:
            print('ERROR divide by zero')
            return
    def product(self):
        return self.a*self.b
    def power(self):
        return self.a**self.b
    def mod(self):
        return self.a%self.b
    def set_numbers(self,a,b):
        self.a=self.b=b
        
a,b=map(int,input().split())
cal=Calculator(a,b)
ans=''
while(ans.lower()!='yes'):
    s=input('Moi nhap phep tinh hoac doi 2 so')
    match s:
        case '+':
            print(cal.add())
        case '-':
            print(cal.subtract())
        case '*':
            print(cal.product())
        case '^':
            print(cal.power())
        case '%':
            print(cal.mod())
        case '/':
            print(cal.divide())
        case _:
            a,b=map(int,s.split())
            cal.set_numbers(a,b)
    ans=input("Dung lai khong ? yes/no")
        
```

```python
#W8A5
class shoppingCart:
    def __init__(self) -> None:
        self.dic={}
        self.cart=[]
    def add(self,name,price):
        self.dic[name]=int(price)
        self.cart.append(name)
    def delete(self,name):
        if name in self.cart:
            self.cart.remove(name)
    def check_cart(self):
        if len(self.cart)==0:
            print('empty')
        else:
            print('not empty')
    def bill(self):
        print(sum(self.dic[i] for i in self.cart))
    def show_cart(self):
        for i in self.cart:
            print(f'{i} :{self.dic[i]}')
    def clear(self):
        self.cart.clear()

cart=shoppingCart()
ans=''
while ans.lower()!='yes':
    s=int(input(""" Mời lựa chọn từ 1 đến 7:
                        2. Thêm vào giỏ hàng một sản phẩm mới
                        3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên
                        để xóa)
                        4. Kiểm tra xem giỏ hàng có rỗng không
                        5. Tính tổng tiền giỏ hàng
                        6. Hiển thị toàn bộ sản phẩm trong giỏ
                        7. Xóa toàn bộ giỏ hàng"""))
    match s:
        case 2:
            a,b=input('Nhập tên và giá sản phẩm').split()
            cart.add(a,b)
        case 3:
            a=input('Nhập tên sản phẩm cần xóa')
            cart.delete(a)
        case 4:
            cart.check_cart()
        case 5:
            cart.bill()
        case 6:
            cart.show_cart()
        case 7:
            cart.clear()
    ans=input("dừng lại không ? yes/no")

    
        
```

```python
#W8A6
class Rectangle:
    def __init__(self,w,h,k):
        self.w=w
        self.h=h
        self.k=k
    def area(self):
        return self.h*self.w
    def perimeter(self):
        return 2*(self.h+self.w)
    def scale(self):
        self.h*=self.k
        self.w*=self.k
w,h,k=map(int,input().split())
rectangle=Rectangle(w,h,k)
rectangle.scale()
print(f'{rectangle.area():.2f} {rectangle.perimeter():.2f}')
```

```python
#W8A7
def gcd(a,b):
    while a:
        a,b=b%a,a
    return b
class Fraction:
    def __init__(self,a,b) -> None:
        a,b=int(a),int(b)
        if b<0:
            b=-b
            a=-a
        d=gcd(a,b)
        self.a=a//d
        self.b=b//d
    def add(self,x):
        self.a=self.a*x.b+x.a*self.b
        self.b=self.b*x.b
    def sub(self,x):
        self.a=self.a*x.b-x.a*self.b
        self.b=self.b*x.b
    def div(self,x):
        if x.a==0 or x.b==0:
            raise RuntimeError ("Error div by 0")
        self.a*=x.b
        self.b*=x.a
    def mul(self,x):
        self.a*=x.a
        self.b*=x.b
    def __str__(self):
        d=gcd(self.a,self.b)
        self.a//=d
        self.b//=d
        if self.b<0:
            self.b=-self.b
            self.a=-self.a
        return f"{self.a}/{self.b}"
s=list(input().split())
s1=Fraction(s[0],s[1])
s2=Fraction(s[3],s[4])
match s[2]:
    case '+':
        s1.add(s2)
    case '-':
        s1.sub(s2)
    case '*':
        s1.mul(s2)
    case '/':
        s1.div(s2)
print(s1)
        
```

```python
#W8A8
class BankAccount:
    def __init__(self,owner,balance) -> None:
        self.owner=owner
        self.balance=int(balance)
    def deposit(self,x):
        self.balance+=x;
    def withdraw(self,x):
        if self.balance>=x:
            self.balance-=x;
    def __str__(self) -> str:
        return f'{self.balance:.2f}'
    
owner,balance=input().split()
p1=BankAccount(owner,balance)
n=int(input())
for _ in range(n):
    cmd,num=input().split()
    if cmd=="DEPOSIT":
        p1.deposit(int(num))
    else:
        p1.withdraw(int(num))
print(p1)
```

```python
#W8A9
class Student:
    def __init__(self,name) -> None:
        self.name=name
        self.sub={}
    def add_score(self,subject,score):
        self.sub[subject]=score
    def avg(self):
        return f'{sum(i for i in self.sub.values())/len(self.sub):.2f}'
    def rank(self):
        s=float(self.avg())
        if s>=8:
            return "Excellent"
        elif s>=6.5:
            return 'Good'
        elif s>=5:
            return 'Average'
        else:
            return "Poor"
    def __str__(self) -> str:
        return f'{self.name} {self.avg()} {self.rank()}'
    
name=input()
s1=Student(name)
n=int(input())
for _ in range(n):
    sub,score=input().split()
    s1.add_score(sub,float(score))
print(s1)
```

```python
#W8A10
class Book:
    def __init__(self,title,author,year) -> None:
        self.title=title
        self.author=author
        self.year=year
class Library:
    def __init__(self) -> None:
        self.bookshelf=[]
    def add(self,book):
        self.bookshelf.append(book)
    def count_by_author(self,name):
        cnt=0
        for i in self.bookshelf:
            if i.author==name:
                cnt+=1
        return cnt
    def find_by_year(self,year):
        cnt=0
        for i in self.bookshelf:
            if i.year==year:
                cnt+=1
        return cnt

l1=Library()
n=int(input())
for _ in range(n):
    cmd,x=input().split()
    if cmd=="ADD":
        title,author,year=x.split(';')
        l1.add(Book(title,author,year))
    elif cmd=='COUNT':
        print(l1.count_by_author(x))
    else:
        print(l1.find_by_year(x))
```

```python
#W8A11
import math
class Point:
    def __init__(self,x,y,z) -> None:
        self.x=x;
        self.y=y
        self.z=z
class math_vector:
    def sub(self,a,b):
        return Point(a.x-b.x,a.y-b.y,a.z-b.z)
    def add(self,a,b):
        return Point(a.x+b.x,a.y+b.y,a.z+b.z)
    def normal_vec(self,a,b,c):
        u,v=self.sub(b,a),self.sub(c,a)
        return Point(u.y*v.z-u.z*v.y,u.z*v.x-u.x*v.z,u.x*v.y-u.y*v.x)
    def dot_product(self,u,v):
        return u.x*v.x+u.y*v.y+u.z*v.z    
    def length(self,u):
        return math.sqrt(self.dot_product(u,u))
    def angle(self,u,v):
        cos_a=abs(self.dot_product(u,v))/(self.length(u)*self.length(v))
        return math.degrees(math.acos(cos_a))
M=math_vector()
point=[]
for _ in range(4):
    x,y,z=map(float,input().split())
    point.append(Point(x,y,z))
a,b,c,d=point
u=M.normal_vec(a,b,c)
v=M.normal_vec(b,c,d)
print(f'{M.angle(u,v):.2f}')

        
```

```python
#W8A12
class Complex_num:
    def __init__(self,a,b) -> None:
        self.re=a
        self.im=b
    def product(self,y):
        a=self.re
        b=self.im
        return Complex_num(a*y.re-b*y.im,b*y.re+a*y.im)
    def __str__(self) -> str:
        return f'{self.re} {self.im}'
num=[]
for _ in range(2):
    a,b=map(float,input().split())
    num.append(Complex_num(a,b))
print(num[0].product(num[1]))
```

```python
#W8A13
class Triangle:
    def __init__(self,a,b,c) -> None:
        self.a=a
        self.b=b
        self.c=c
    def check(self):
        a,b,c=sorted([self.a,self.b,self.c])
        return a+b>c
    def getArea(self):
        a,b,c=self.a,self.b,self.c
        if not self.check():
            return('Invalid')
        else:
            p=(a+b+c)/2
            return f'{(p*(p-a)*(p-b)*(p-c))**0.5:.2f}'
a,b,c=map(int,input().split())
t1=Triangle(a,b,c)
print(t1.getArea())
```
