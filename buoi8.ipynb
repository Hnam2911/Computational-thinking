```python
#W8A1
import math
pi=3.14
class Cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def total_area(self):
        return 2*pi*((self.r)**2)+2*(self.r)*pi*self.h
    def volume(self):
        return (self.r**2)*pi*self.h
r,h=map(int,input().split())
cy1=Cylinder(r,h)
print(cy1.volume())
print(cy1.total_area())
```

```python
#W8A3
class Point:
    def __init__(self,x,y) -> None:
        self.x=x
        self.y=y
    def check_0(self):
        return self.x==self.y==0
    def check_x_axis(self):
        return self.x==0
    def check_y_axis(self):
        return self.y==0
    def dis(self,a):
        res=((self.x-a.x)**2+(self.y-a.y)**2)**0.5
        return f'{res:.2f}'
x,y=map(int,input().split())
p1=Point(x,y)
print(p1.check_0())
print(p1.check_x_axis())
print(p1.check_y_axis())
print(p1.dis(Point(0,0)))
```

```python
#W8A4
class Calculator:
    def __init__(self,a,b) -> None:
        self.a=a
        self.b=b
    def add(self):
        return self.a+self.b
    def subtract(self):
        return self.a-self.b
    def divide(self):
        if self.b!=0:
            return self.a/self.b
        else:
            print('ERROR divide by zero')
            return
    def product(self):
        return self.a*self.b
    def power(self):
        return self.a**self.b
    def mod(self):
        return self.a%self.b
    def set_numbers(self,a,b):
        self.a=self.b=b
        
a,b=map(int,input().split())
cal=Calculator(a,b)
ans=''
while(ans.lower()!='yes'):
    s=input('Moi nhap phep tinh hoac doi 2 so')
    match s:
        case '+':
            print(cal.add())
        case '-':
            print(cal.subtract())
        case '*':
            print(cal.product())
        case '^':
            print(cal.power())
        case '%':
            print(cal.mod())
        case '/':
            print(cal.divide())
        case _:
            a,b=map(int,s.split())
            cal.set_numbers(a,b)
    ans=input("Dung lai khong ? yes/no")
        
```

```python
#W8A5
class shoppingCart:
    def __init__(self) -> None:
        self.dic={}
        self.cart=[]
    def add(self,name,price):
        self.dic[name]=int(price)
        self.cart.append(name)
    def delete(self,name):
        if name in self.cart:
            self.cart.remove(name)
    def check_cart(self):
        if len(self.cart)==0:
            print('empty')
        else:
            print('not empty')
    def bill(self):
        print(sum(self.dic[i] for i in self.cart))
    def show_cart(self):
        for i in self.cart:
            print(f'{i} :{self.dic[i]}')
    def clear(self):
        self.cart.clear()

cart=shoppingCart()
ans=''
while ans.lower()!='yes':
    s=int(input(""" Mời lựa chọn từ 1 đến 7:
                        2. Thêm vào giỏ hàng một sản phẩm mới
                        3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên
                        để xóa)
                        4. Kiểm tra xem giỏ hàng có rỗng không
                        5. Tính tổng tiền giỏ hàng
                        6. Hiển thị toàn bộ sản phẩm trong giỏ
                        7. Xóa toàn bộ giỏ hàng"""))
    match s:
        case 2:
            a,b=input('Nhập tên và giá sản phẩm').split()
            cart.add(a,b)
        case 3:
            a=input('Nhập tên sản phẩm cần xóa')
            cart.delete(a)
        case 4:
            cart.check_cart()
        case 5:
            cart.bill()
        case 6:
            cart.show_cart()
        case 7:
            cart.clear()
    ans=input("dừng lại không ? yes/no")

    
        
```

```python
#W8A6
class Rectangle:
    def __init__(self,w,h,k):
        self.w=w
        self.h=h
        self.k=k
    def area(self):
        return self.h*self.w
    def perimeter(self):
        return 2*(self.h+self.w)
    def scale(self):
        self.h*=self.k
        self.w*=self.k
w,h,k=map(int,input().split())
rectangle=Rectangle(w,h,k)
rectangle.scale()
print(f'{rectangle.area():.2f} {rectangle.perimeter():.2f}')
```

```python
#W8A7
def gcd(a,b):
    while a:
        a,b=b%a,a
    return b
class Fraction:
    def __init__(self,tu,mau) -> None:
        tu,mau=int(tu),int(mau)
        if mau<0:
            mau=-mau
            tu=-tu
        a=gcd(tu,mau)
        self.tu=tu//a
        self.mau=mau//a
    def add(self,x):
        tu=self.tu*x.mau+x.tu*self.mau
        mau=self.mau*x.mau
        a=gcd(tu,mau)
        self.tu=tu//a
        self.mau=mau//a
        if self.mau<0:
            self.mau=-self.mau
            self.tu=-self.tu
    def sub(self,x):
        tu=self.tu*x.mau-x.tu*self.mau
        mau=self.mau*x.mau
        a=gcd(tu,mau)
        self.tu=tu//a
        self.mau=mau//a
        if self.mau<0:
            self.mau=-self.mau
            self.tu=-self.tu
    def div(self,x):
        self.tu*=x.mau
        self.mau*=x.tu
        a=gcd(self.tu,self.mau)
        self.tu//=a
        self.mau//=a
        if self.mau<0:
            self.mau=-self.mau
            self.tu=-self.tu
    def mul(self,x):
        self.tu*=x.tu
        self.mau*=x.mau
        a=gcd(self.tu,self.mau)
        self.tu//=a
        self.mau//=a
        if self.mau<0:
            self.mau=-self.mau
            self.tu=-self.tu
    def __str__(self):
        return f"{self.tu}/{self.mau}"
s=list(input().split())
s1=Fraction(s[0],s[1])
s2=Fraction(s[3],s[4])
match s[2]:
    case '+':
        s1.add(s2)
    case '-':
        s1.sub(s2)
    case '*':
        s1.mul(s2)
    case '/':
        s1.div(s2)
print(s1)
        
```
